```python
from keras.layers import Input, Dense
from keras.models import Model

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

np.random.seed(42)
n_samples = 1000
X = np.random.rand(n_samples, 10)  # 10 feature columns
y = np.random.randint(0, 2, size=(700,))
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define the autoencoder architecture
input_layer = Input(shape=(X_train.shape[1],))
encoded = Dense(64, activation='relu')(input_layer)
decoded = Dense(X_train.shape[1], activation='sigmoid')(encoded)

autoencoder = Model(input_layer, decoded)

# Compile and train the autoencoder
autoencoder.compile(optimizer='adam', loss='mean_squared_error')
autoencoder.fit(X_train, X_train, epochs=50, batch_size=256, shuffle=True)

# Use the encoder to reduce the feature set
encoder = Model(input_layer, encoded)
encoded_features = encoder.predict(X_train)
```

```python
print(encoded_features)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
encoded_features_reshaped = np.expand_dims(encoded_features, axis=-1)  # shape (1000, 50, 1, 1)

# Now, the data is in the shape that the CNN expects: (samples, height, width, channels)

# Build a simple CNN model for this 1D feature data
# Build the CNN model
model = Sequential()

# First Conv2D layer with kernel size (3, 1) for 1D-like data
model.add(Conv2D(64, kernel_size=(3, 1), activation='relu', input_shape=(64, 1, 1)))

# Max pooling layer
model.add(MaxPooling2D(pool_size=(2, 1)))

# Flatten the output from Conv2D
model.add(Flatten())

# Now, let's check the shape of the output after Flatten
model.summary()

# Add Dense layer
model.add(Dense(128, activation='relu'))  # Example for fully connected layer
model.add(Dense(1, activation='sigmoid'))  # Binary classification output layer

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Example labels
labels = np.random.randint(0, 2, size=(1000,))  # Example binary labels

# Train the model
model.fit(encoded_features, labels, epochs=10, batch_size=32)
```
